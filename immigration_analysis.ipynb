```
import pandas as pd
```

```
df = pd.read_csv("/content/imigrantes_canada.csv")
```

```
df
```

```
df.info()
```

```
df.set_index("País", inplace=True)
```

```
year = list(map(str, range(1980, 2014)))
```

```
brazil = df.loc["Brasil", year]
```

```
brazil
```

```
brazil_dict = {"Year": brazil.index.tolist(), "Immigrants": brazil.values.tolist()}
data_brazil = pd.DataFrame(brazil_dict)
```

```
data_brazil
```

```
import matplotlib.pyplot as plt
```

```
plt.figure(figsize=(8,4))
plt.plot(data_brazil["Year"], data_brazil["Immigrants"])
plt.xlabel("Year")
plt.ylabel("Immigrants")
plt.title("Immigration from Brazil to Canada")
plt.xticks(["1980", "1985", "1990", "1995", "2000", "2005", "2010"])
plt.show()
```
